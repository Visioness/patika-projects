```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('survey_results_public.csv')
```

```python
df = data[['ResponseId', 'MainBranch', 'Age', 'Employment', 'RemoteWork',
'CodingActivities', 'EdLevel', 'LearnCode', 'LearnCodeOnline', 'LearnCodeCoursesCert',
'YearsCode', 'YearsCodePro', 'DevType', 'OrgSize', 'BuyNewTool', 'Country', 'Currency', 'CompTotal',
'LanguageHaveWorkedWith', 'DatabaseHaveWorkedWith', 'PlatformHaveWorkedWith', 'WebframeHaveWorkedWith',
'MiscTechHaveWorkedWith', 'ToolsTechHaveWorkedWith', 'NEWCollabToolsHaveWorkedWith', 'OpSysPersonal use',
'OpSysProfessional use', 'AISearchHaveWorkedWith', 'AIDevHaveWorkedWith']]
```

```python
df.shape
```

```python
df.isna().sum().to_frame()
```

```python
columns = ['CodingActivities', 'LearnCode', 'LearnCodeOnline', 'LearnCodeCoursesCert', 'BuyNewTool',
'LanguageHaveWorkedWith', 'DatabaseHaveWorkedWith', 'PlatformHaveWorkedWith', 'WebframeHaveWorkedWith', 'MiscTechHaveWorkedWith',
'ToolsTechHaveWorkedWith', 'NEWCollabToolsHaveWorkedWith', 'OpSysPersonal use', 'OpSysProfessional use', 'AISearchHaveWorkedWith', 'AIDevHaveWorkedWith']

stacked = {}

for col in columns:
	stacked[col] = df[col].str.split(';', expand=True).stack()
```

```python
num_learncode = stacked['LearnCode'].value_counts()
```

```python
sns.set_theme(palette='mako')
plt.figure(figsize=(14, 8))

sns.barplot(x=num_learncode.values, y=num_learncode.index, orient='h')
plt.xlabel('Individiual Count', fontsize=14, fontweight='bold')
plt.ylabel('General Categories (Multiple Choice)', fontsize=14, fontweight='bold')
plt.xticks(rotation=30)
plt.title('Where did you learn to code?', fontsize=18, fontweight='bold')

plt.show()
```

```python
num_learncode_online = stacked['LearnCodeOnline'].value_counts()
```

```python
sns.set_theme(palette='mako')
plt.figure(figsize=(14, 8))

sns.barplot(x=num_learncode_online.values, y=num_learncode_online.index, orient='h')
plt.xlabel('Individiual Count', fontsize=14, fontweight='bold')
plt.ylabel('Online Categories (Multiple Choice)', fontsize=14, fontweight='bold')
plt.xticks(rotation=30)
plt.title('Where did you learn to code?', fontsize=18, fontweight='bold')

plt.show()
```

```python
num_learncode_courses = stacked['LearnCodeCoursesCert'].value_counts()
```

```python
sns.set_theme(palette='mako')
plt.figure(figsize=(14, 8))

sns.barplot(x=num_learncode_courses.values, y=num_learncode_courses.index, orient='h')
plt.xlabel('Individiual Count', fontsize=14, fontweight='bold')
plt.ylabel('Course Categories (Multiple Choice)', fontsize=14, fontweight='bold')
plt.xticks(rotation=30)
plt.title('Where did you learn to code?', fontsize=18, fontweight='bold')

plt.show()
```

```python
df_by_countries = df.groupby('Country')['Country'].value_counts().sort_values(ascending=False)
df_by_countries[:15]
```

```python
sns.set_theme(palette='mako')
plt.figure(figsize=(14, 8))

sns.barplot(x=df_by_countries.values[:15], y=df_by_countries.index[:15], orient='h')
plt.xlabel('Individiual Count', fontsize=14, fontweight='bold')
plt.ylabel('Countries', fontsize=14, fontweight='bold')
plt.xticks(rotation=30)
plt.title('Where are you from?', fontsize=18, fontweight='bold')

plt.show()
```

```python
df_by_age = df.groupby('Age')['Age'].value_counts()
```

```python
sns.set_theme(palette='mako')
plt.figure(figsize=(14, 8))

sns.barplot(x=df_by_age.values, y=df_by_age.index, order=['65 years or older', '55-64 years old', '45-54 years old', '35-44 years old', '25-34 years old', '18-24 years old', 'Under 18 years old', 'Prefer not to say'] ,orient='h')
plt.xlabel('Individiual Count', fontsize=14, fontweight='bold')
plt.ylabel('Age', fontsize=14, fontweight='bold')
plt.xticks(rotation=30)
plt.title('How old are you?', fontsize=18, fontweight='bold')

plt.show()
```
